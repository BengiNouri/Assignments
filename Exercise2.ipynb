## Experiment with Gradiant descent 
Implement a simple gradient descent algorithm to find the minimum value of a quadratic function  f(x)=x2−4x+4 

The goal is to understand how different learning rates affect the convergence towards the function's minimum.

Your task: Manually implement the gradient descent algorithm and observe the convergence process for at least three different learning rates

```python
# First, you need to define the function f(x) = x^2 - 4x + 4 and its derivative.
def f(x):
    return x**2 - 4*x + 4
def df(x):
    return 2*x - 4
print(f(10))    
```

Add the gradient descent update rule  x=x−η⋅f′(x) , where  x  is the current position,  η  is the learning rate and  f′(x)  is the derivative of the function at  x .

```python
def gradient_descent(initial_x, learning_rate, epochs):
    x = initial_x
    history = [] # to store value of x and f(x) at each step
    for i in range(epochs):
        gradient = df(x)
        x = x - learning_rate * gradient
        history.append((x, f(x)))
        return history
```

It's time to perform gradient descent for some learning rates. First, try the learning rates below. Then try higher learning rates. What happens at higher rates? Why?

```python
import matplotlib.pyplot as plt

learning_rates = [0.01, 0.1, 0.5]
initial_x = 0  # Starting point
epochs = 10

for lr in learning_rates:
    history = gradient_descent(initial_x, lr, epochs)
    xs, ys = zip(*history)
    plt.plot(xs, label=f'LR={lr}')

plt.xlabel('Iteration')
plt.ylabel('x value')
plt.title('Gradient Descent Convergence')
plt.legend()
plt.show()
```
